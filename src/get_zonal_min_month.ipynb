```python
import pandas as pd
import numpy as np
from pathlib import Path

BASEDIR = Path('.').resolve().parent
df = pd.read_csv(str(BASEDIR.joinpath('dataout/HTI/HTI_zonal_stats_National.csv')))
cols = []
for i in range(12):
    i += 1
    if len(str(i)) < 2:
        month = f'sum0{str(i)}'
    else:
        month = f'sum{str(i)}'
    cols.append(month)
#print(cols)
#print(df)
df = df[cols].min()
df.idxmin(axis=1)
```

```python
country = 'HTI'
min_month = 'sum04'
df = pd.read_csv(BASEDIR.joinpath(f'dataout/{country}/{country}_zonal_stats.csv'))
cols = []
for i in range(12):
    i += 1
    if len(str(i)) < 2:
        month = f'sum0{str(i)}'
    else:
        month = f'sum{str(i)}'
    cols.append(month)
for i in cols:
    
    df.insert(1, f'SoL_obs{i[-2:]}', value=df[i]/df[min_month])
df = df.fillna(value=0)
df.columns.values()
```
